```python
url = "https://worldpopulationreview.com/zips/colorado"
web_list = pd.read_html(url)
pop_df = web_list[0]
pop_df.head()
```

```python
web_list
```

```python
merged_df = clean_fuel.merge(pop_df, left_on='zip', right_on='Zip Code')
merged_df
```

```python
income_list = []
for i in range(1, 6):
    if i == 1:
        url_3 = 'http://zipatlas.com/us/co/zip-code-comparison/median-household-income.htm'
        income_list.append(pd.read_html(url_3)[10])   
    else:    
        url_3 = 'http://zipatlas.com/us/co/zip-code-comparison/median-household-income.'+str(i) + '.htm'
        new_list = (pd.read_html(url_3)[10])
        new_list = new_list[1:]
        income_list.append(new_list)
        

income_df = pd.concat(income_list)
# Adjust column header
new_header = income_df.iloc[0] #grab the first row for the header
income_df = income_df[1:] #take the data less the header row
income_df.columns = new_header #set the header row as the df header
income_df.head()
```

```python
income_df = income_df[['Zip Code', 'Avg. Income/H/hold']]
income_df['Zip Code'] = income_df['Zip Code'].apply(pd.to_numeric)
income_df
```

```python
merged_df2 = merged_df.merge(income_df, left_on='Zip Code', right_on='Zip Code')
final_df = merged_df2.drop(columns=['zip', 'city'], axis=1 )
final_df.head()
```

```python
final_df['Avg. Income/H/hold'] = final_df['Avg. Income/H/hold'].str.replace('$', '')
final_df['Avg. Income/H/hold'] = final_df['Avg. Income/H/hold'].str.replace(',', '')
final_df['Avg. Income/H/hold'] = final_df['Avg. Income/H/hold'].astype(float)
final_df = final_df.rename(columns={'Avg. Income/H/hold': 'Average Household Income'})
final_df
```
